```python
create_table_query = """
CREATE TABLE IF NOT EXISTS accounts_country (
    code VARCHAR primary key,
    short_name VARCHAR,
    table_name VARCHAR,
    long_name VARCHAR,
    currency_unit VARCHAR
);
"""
cur.execute(create_table_query)


conn.commit()
```

```python
create_table_query="""
CREATE TABLE IF NOT EXISTS accounts_data (
country_name VARCHAR,
country_code VARCHAR,
series_name VARCHAR,
series_code VARCHAR,
year_1995 VARCHAR,
year_2000 VARCHAR,
year_2005 VARCHAR,
year_2010 VARCHAR,
year_2014 VARCHAR
);
"""

cur.execute(create_table_query)

conn.commit()
```

```python
create_table_query="""
CREATE TABLE IF NOT EXISTS accounts_series(
code VARCHAR,
topic VARCHAR,
indicator_name VARCHAR,
long_definition VARCHAR);
"""


cur.execute(create_table_query)

conn.commit()
```

```python

accounts_country_insert_into_query = """
INSERT INTO accounts_country(
    code,
    short_name,
    table_name,
    long_name,
    currency_unit
)
VALUES(%s, %s, %s, %s, %s)ON CONFLICT (Code)DO NOTHING;
"""

for i, row in AccountsCountry_clean.iterrows():
    data_row = (
        row['Code'],
        row['Short Name'],
        row['Table Name'],
        row['Long Name'],
        row['Currency Unit']
    )
    print(data_row)
    cur.execute(accounts_country_insert_into_query, data_row)


conn.commit()
```

```python
accounts_series_insert_into_query="""
INSERT INTO accounts_series(
    code,
    topic,
    indicator_name,
    long_definition
)VALUES(%s, %s, %s, %s);
"""


for i,row in AccountsSeries.iterrows():
    data_row=(
        row['Code'],
        row['Topic'],
        row['Indicator Name'],
        row['Long definition']
    )
    print(data_row)
    cur.execute(accounts_series_insert_into_query,data_row)

conn.commit()
```
